### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (Top-Left)
axes[0, 0].plot(x, y, color='blue', label='y = x^2')
axes[0, 0].set_title("Line Plot")
axes[0, 0].legend()
axes[0, 0].grid(True)

# Bar plot (Top-Right)
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")

# Scatter plot (Bottom-Left)
axes[1, 0].scatter(scatter_x, scatter_y, color='blue')
axes[1, 0].set_title("Scatter Plot")
for i in range(len(scatter_x)):
    axes[1, 0].text(scatter_x[i], scatter_y[i], f"({scatter_x[i]}, {scatter_y[i]})", fontsize=9, ha='right')

# Pie chart (Bottom-Right)
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, shadow=True)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
